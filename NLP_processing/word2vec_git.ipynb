```python
import gensim
# Skip gram model 로드
# Load Google's pre-trained Word2Vec model.
#https://drive.google.com/file/d/0B7XkCwpI5KDYNlNUTTlSS21pQmM/edit 에서 googleNews vector 다운로드
# 다운로드한 파일을 NLP root 아래에 위치시키기
NLP_dir='자신의 NLP_root_dir 설정'
model = gensim.models.KeyedVectors.load_word2vec_format(NLP_dir,'/GoogleNews-vectors-negative300.bin', binary=True)  
```

```python
print(len(model.vocab))
```

```python
voca=model.vocab
word=list(voca.keys())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import codecs
from operator import eq  #exactly search 하는 패키지
```

```python
# skip gram pretrained word 저장 하기 
word_list = codecs.open(NLP_dir,'/skipgram/skipgram_word_list.txt','w',"utf-8")
for s in word:
    word_list.write(str(s)+'\n')
word_list.close()
```

```python
#중복제거된 source description, target description영문 불러오기
source_en_list = pd.read_csv(NLP_dir,'/description/840B/중복제거/source/save_source_en.txt',header = None, delimiter="\t", quoting=csv.QUOTE_NONE, encoding='utf-8')
source_en_list=source_en_list[0]
target_en_list = pd.read_csv(NLP_dir,'/description/840B/중복제거/target/save_target_en.txt',header = None, delimiter="\t", quoting=csv.QUOTE_NONE, encoding='utf-8')
target_en_list=target_en_list[0]

#중복 허용된 영문 total description 불러오기 
total_dup_en_list = pd.read_csv(NLP_dir,'/description/840B/중복허용/source_target/save_en_total_list_org.txt',header = None, delimiter="\t", quoting=csv.QUOTE_NONE, encoding='utf-8')
total_dup_en_list=total_dup_en_list[0]
```

```python
#source decription이 포함되는 단어인지 아닌지 list로 만들어 저장
inner_source_skipgram = []
outter_source_skipgram =[]
for e_list in source_en_list:
    for i,b_corpus in enumerate(word):
        if eq(e_list,b_corpus) :
            inner_source_skipgram = inner_source_skipgram + [e_list]
            break
        elif i == len(word)-1:
            outter_source_skipgram = outter_source_skipgram + [e_list]

print('원래 영문 source list 길이: ',len(source_en_list))
print('skipgram에 속한 source decription영문 개수: ',len(inner_source_skipgram))
print('skipgram에 속하지 않은 source decription영문 개수: ',len(outter_source_skipgram))

#
save_token = codecs.open(NLP_dir,'/description/skipgram/source/inner_source_skipgram_ENlist.txt','w',"utf-8")
for s in inner_source_skipgram:
    save_token.write(str(s)+'\n')
save_token.close()

#
save_token2 = codecs.open(NLP_dir,'/description/skipgram/source/outter_source_skipgram_ENlist.txt','w',"utf-8")
for s in outter_source_skipgram:
    save_token2.write(str(s)+'\n')
save_token2.close()
```

```python
#source, description 중복허용하여 단어별로 정리한 파일 
total_ = pd.read_csv(NLP_dir,'/description/840B/중복허용/source_target/save_en_total_list_org.txt',header = None, delimiter="\t", quoting=csv.QUOTE_NONE, encoding='utf-8')
total_ = total_[0]
print(len(total_))
```

```python
# 중복 total list에 있는 값을 map화 시켜 count 
from collections import Counter
total_ = list(total_)
print("---Counter()---")
result = Counter(total_)
print(result)
```

```python
#타겟 결과 카운트
target_outter_dict={}
for v in outter_target_skipgram:
    target_outter_dict[''+str(v)]=result[''+str(v)]
    
print(len(target_outter_dict))
# sorted(target_outter_dict, key=target_outter_dict.get)
outter_target_count = sorted(target_outter_dict.items(), key=lambda kv: kv[1])
# csv로 outter target file 저장
outter_target_count
dataframe = pd.DataFrame(outter_target_count)
dataframe.to_csv(NLP_dir,'/description/skipgram/count파일/outter_target/outter_target_count.csv',header=False,index=False)
```

```python
#타겟 결과 카운트
source_outter_dict={}
for v in outter_source_skipgram:
    source_outter_dict[''+str(v)]=result[''+str(v)]
    
print(len(source_outter_dict))
# sorted(target_outter_dict, key=target_outter_dict.get)
outter_source_count = sorted(source_outter_dict.items(), key=lambda kv: kv[1])
# csv로 outter target file 저장
outter_source_count
dataframe = pd.DataFrame(outter_source_count)
dataframe.to_csv(NLP_dir,'/description/skipgram/count파일/outter_source/outter_source_count.csv',header=False,index=False)
```
